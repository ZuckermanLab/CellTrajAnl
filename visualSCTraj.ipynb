```python
import tifffile as tiff

# Load the TIFF file
with tiff.TiffFile('VID829_A1_1_00d00h00m.tif') as tif_file:
    # Display general information about the TIFF file
    print(tif_file)
    
    # Get image data
    image_data = tif_file.asarray()
    
    # Display the shape and data type of the image data
    print(f"Image shape: {image_data.shape}")
    print(f"Image data type: {image_data.dtype}")
    
    # If you want more detailed information about each page (especially for multi-page TIFFs)
    for page in tif_file.pages:
        print(page.tags)

# Optional: Display the image using matplotlib
import matplotlib.pyplot as plt
plt.imshow(image_data, cmap='gray')
plt.title("TIFF Image")
plt.axis('off')
plt.show()
```
