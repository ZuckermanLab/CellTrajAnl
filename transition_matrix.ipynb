```python
import matplotlib.pyplot as plt
import numpy as np
from joblib import dump, load
```

```python
fig, ax = plt.subplots(figsize = (7, 6))

#filename = 'transitionMatrix_LI204601_P_40_Sep07-2023Xpc10.joblib'
#filename = 'transitionMatrix_LI204601_P_40_Sep08-2023.joblib'
filename = 'tMat_LI204601_P_40_Sep11-2023pc3umap2AllWells'
P = load(filename+'.joblib')
P = np.array(P, dtype=np.float64)
plt.title('Transition Matrix: 24 hrs duration, 10 hrs snippet length', fontsize = 8)
im = ax.imshow(P)
ax.set_xticks(np.arange(0, P.shape[1], 20))
ax.set_yticks(np.arange(0, P.shape[0], 20))
ax.set_ylim(-0.5, P.shape[0]-0.5)
ax.set_ylim(ax.get_ylim()[::-1])
   
#adding an axes, lower left corner in [0.91, 0.1] measured in figure coordinate with axes width 0.02 and height 0.8
cb_ax = fig.add_axes([0.91, 0.1, 0.02, 0.75])
cbar = fig.colorbar(im, cax = cb_ax)
#plt.tight_layout()
#plt.savefig('transitionMatrix_LI204601_P_40_Sep07-2023Xpc10.png', dpi = 150, bbox_inches='tight')
plt.savefig(filename+'.png', dpi = 150, bbox_inches='tight')
plt.show()
```

```python
trajectory_lengths = [8, 16, 24, 32, 40, 48]
ntrajLen = len(trajectory_lengths)

fig, ax = plt.subplots(2, 3, figsize = (12, 9))

for i in range(ntrajLen):
    if i < 3:
        i1 = 0; i2 = i
    else:
        i1 = 1; i2 = 3-i
    filename = 'transitionMatrix_LI204601_P_'+str(trajectory_lengths[i])+'.joblib'
    #print(filename)
    P = load(filename)
    P = np.array(P, dtype=np.float64)
    plt.subplot(2, 3, i+1)
    plt.title('tLen = '+str(trajectory_lengths))
    im = ax[i1][i2].imshow(P)
    ax[i1][i2].set_xticks(np.arange(0, P.shape[1], 20))
    ax[i1][i2].set_yticks(np.arange(0, P.shape[0], 20))
    ax[i1][i2].set_ylim(-0.5, P.shape[0]-0.5)
    ax[i1][i2].set_ylim(ax[i1][i2].get_ylim()[::-1])
   
# adding an axes, lower left corner in [0.91, 0.1] measured in figure coordinate with axes width 0.02 and height 0.8
cb_ax = fig.add_axes([0.91, 0.1, 0.02, 0.8])
cbar = fig.colorbar(im, cax = cb_ax)
#plt.tight_layout()
plt.show()
```
